```python
import os
from tqdm import tqdm
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

## Carotid mutinfo VAL

```python
moved_frames_file = open('moved_MOSTLY_black_frames_CAROTID_MUTINFO_VAL.txt', 'a')

VOLUME_FOLDER="C:/Users/Tmenova/personal/tum/thesis/initial-data/new_datasets/CAROTID_MUTINFO/val/images"
LABELS_FOLDER="C:/Users/Tmenova/personal/tum/thesis/initial-data/new_datasets/CAROTID_MUTINFO/val/labels"
REMOVED_IMAGES_FOLDER="C:/Users/Tmenova/personal/tum/thesis/initial-data/new_datasets/CAROTID_MUTINFO/removed2/val/images"
REMOVED_LABELS_FOLDER="C:/Users/Tmenova/personal/tum/thesis/initial-data/new_datasets/CAROTID_MUTINFO/removed2/val/labels"

if not os.path.exists(REMOVED_IMAGES_FOLDER):
    os.makedirs(REMOVED_IMAGES_FOLDER)

if not os.path.exists(REMOVED_LABELS_FOLDER):
    os.makedirs(REMOVED_LABELS_FOLDER)

```

```python
v_slice_file_path = os.path.join(VOLUME_FOLDER, "ultrasonix_test254.png")
v = np.array(Image.open(v_slice_file_path).convert('L'))
plt.imshow(v)
```

```python
num_non_zero = np.count_nonzero(v > 1)
ratio_nz = num_non_zero/v.size
ratio_nz
```

```python
PERCENTAGE_THRESHOLD=0.6
```

```python
# moved frames that are MOSTLY black
for v_slice_file, segm_slice_file in zip(sorted(os.listdir(VOLUME_FOLDER)), sorted(os.listdir(LABELS_FOLDER))):
# for v_slice_file in sorted(os.listdir(VOLUME_FOLDER)):
    v_slice_file_path = os.path.join(VOLUME_FOLDER, v_slice_file)
    v = np.array(Image.open(v_slice_file_path).convert('L'))
    num_non_zero = np.count_nonzero(v > 1)
    ratio_nz = num_non_zero/v.size
    if ratio_nz < PERCENTAGE_THRESHOLD:
        # mostly black rame detected - can move the slice and corresponding segmentation
        v_slice_file_path_moved = os.path.join(REMOVED_IMAGES_FOLDER, v_slice_file)
        os.rename(v_slice_file_path, v_slice_file_path_moved)
        assert(v_slice_file==segm_slice_file)
        s_slice_file_path = os.path.join(LABELS_FOLDER, segm_slice_file)
        s_slice_file_path_moved = os.path.join(REMOVED_LABELS_FOLDER, segm_slice_file)
        os.rename(s_slice_file_path, s_slice_file_path_moved)
        # write in file which frames were deleted
        moved_frames_file.writelines(f'moved volume slice: {v_slice_file_path_moved} \n')
        moved_frames_file.writelines(f'moved segm slice: {s_slice_file_path_moved} \n')

moved_frames_file.close()     

print('images num: ', len(os.listdir(VOLUME_FOLDER)))
print('lebles num: ', len(os.listdir(LABELS_FOLDER)))
print('REMOVED images num: ', len(os.listdir(REMOVED_IMAGES_FOLDER)))
print('REMOVED lebels num: ', len(os.listdir(REMOVED_LABELS_FOLDER)))
```
